```python
from neural_net import *
import requests, pickle
```

```python
with open('./data/mnist1d_data.pkl', 'rb') as handle:
        data = pickle.load(handle)
        
print(data.keys())
```

```python
y_train = [[0 if j != i else 1 for j in range(10)] for i in data['y']]
y_test = [[0 if j != i else 1 for j in range(10)] for i in data['y_test']]
```

```python
adam = Adam()
bce = MSE()
nn = NN(layers=[Linear(data['x'].shape[1], 10, Relu(), init_mode="xavier_uniform"),
                    Linear(10, 25, Relu(), init_mode="xavier_uniform"),
                    Linear(25, 50, Relu(), init_mode="xavier_uniform"),
                    Linear(50, 100, Relu(), init_mode="xavier_uniform"),
                    Linear(100, 50, Relu(), init_mode="xavier_uniform"),
                    Linear(50, 25, Relu(), init_mode="xavier_uniform"),
                    Linear(25, 10, Sigmoid(), init_mode="xavier_uniform"),
                    ],
                    optimizer=adam,
                    loss=bce)

nn.train(data['x'], y_train, data['x_test'], y_test, 100, 10)
```
